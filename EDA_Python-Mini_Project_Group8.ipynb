### 4. Data Cleaning

#### I.a. Identification of Null-values

```python
# 1.1. Overview of nulls per column
null_summary = df.isna().sum()
print(null_summary)
print('\n------------------------------------------------\n')

# 1.2. Percentage of nulls per column
null_pct = (df.isna().mean() * 100).round(2)
print(null_pct)
print('\n------------------------------------------------\n')

# 1.3. Rows with *any* nulls
rows_with_any_nulls = df[df.isna().any(axis=1)]
print("Rows with at least one null:", rows_with_any_nulls.shape[0])
print('\n------------------------------------------------\n')

# 1.4. Rows with *any* nulls having more than 1 null values
rows_with_any_nulls = df[df.isna().any(axis=1) >= 2]
print("Rows with at least one null:", rows_with_any_nulls.shape[0])
print('\n------------------------------------------------\n')

# 1.5. Rows with *all* nulls (unlikely here, but for completeness)
rows_all_nulls = df[df.isna().all(axis=1)]
print("Rows completely empty:", rows_all_nulls.shape[0])
print('\n------------------------------------------------\n')

plt.figure(figsize=(12, 6))
plt.title("Heatmap Showing Missing Values")
sns.heatmap(df.isna(), cbar=False)
plt.show()
```

#### I.b. Handling Null Values

```python
# Check if mode is available before using it
insurance_mode =  df['Insurance Provider'].mode()

if not insurance_mode.empty:
    mode_val = insurance_mode[0]
    df['Insurance Provider'] = df['Insurance Provider'].fillna(mode_val)
else:
    print("No mode found — the column might be entirely null!")
```

```python
# 1. Fill Insurance Provider first (so we preserve mode)
df['Insurance Provider'] = df['Insurance Provider'].fillna(df['Insurance Provider'].mode().iloc[0] \
    if not df['Insurance Provider'].mode().empty \
    else 'Unknown')

# 2. THEN drop rows where both are null
df = df[~(df['Blood Type'].isna() & df['Insurance Provider'].isna())].copy()

# 3. Fill Blood Type
df['Blood Type'] = df['Blood Type'].fillna('Unknown')
```

#### II.a. Check For Duplicates

```python
df.duplicated().sum()
```

```python
# View the duplicate entries (first 5)
df[df.duplicated()].shape
```

```python
# Step 1: Find all duplicated rows (including originals)
dups_df = df[df.duplicated(keep=False)].copy()

# Step 2: Create a unique "signature" of each row to group identical ones
dups_df['row_signature'] = dups_df.apply(lambda row: tuple(row), axis=1)

# Step 3: Map each signature to its first (original) index
signature_to_original_index = {
    sig: dups_df[dups_df['row_signature'] == sig].index.min()
    for sig in dups_df['row_signature'].unique()
}

# Step 4: Print results in requested format
for idx in dups_df.index:
    sig = dups_df.loc[idx, 'row_signature']
    original_idx = signature_to_original_index[sig]
    
    if idx == original_idx:
        print(f"{idx}. Original row")
    else:
        print(f"{idx}. Duplicate of original row {original_idx}")
```

```python
# 1. Create a new column to trace original position
df['order_num'] = df.index+1
```

```python
df.shape
```

#### II.b. Drop the duplicate rows, where handled well since order_num column value is unique. 

```python
df = df.drop_duplicates(subset=[col for col in df.columns if col != 'order_num']).reset_index(drop=True)
```

## 6. Advanced Python Techniques

### I. Lambda Functions

```python
df['Age_Group'] = df['Age'].apply(lambda x: 'Young' if x < 30 else 'Middle-aged' if 30 <= x < 60 else 'Senior')
df['Age_Group'].value_counts()
```

```python
avg_stay = df['Stay_Duration'].mean()
df['Long_Stay'] = df['Stay_Duration'].apply(lambda x: 'Yes' if x > avg_stay else 'No')
df['Long_Stay'].value_counts()
```

```python
df['High_Bill'] = df['Billing_Amount'].apply(lambda x: 1 if x > 30000 else 0)
df['High_Bill'].value_counts()
```

### II. User-Defined Functions

```python
def classify_billing(amount):
    if amount < 15000:
        return 'Low'
    elif 15000 <= amount < 30000:
        return 'Medium'
    else:
        return 'High'

df['Billing_Category'] = df['Billing_Amount'].apply(classify_billing)
df.head(10)
```

```python
def hospital_capacity(room_no):
    if room_no < 150:
        return 'Small'
    elif 150 <= room_no <= 350:
        return 'Medium'
    else:
        return 'Large'

df['Hospital_Size'] = df['Room_Number'].apply(hospital_capacity)
df.head(10)
```

### III. List Comprehensions

```python
# Flag if hospital name contains "Smith"
df['has_smith_hospital'] = ['smith' in str(hosp).lower() for hosp in df['Hospital']]
df['has_smith_hospital'].value_counts()
```

```python
# Flag Gender column to 1/0 --> 
df['is_male'] = [1 if gender == 'Male' else 0 for gender in df['Gender']]
df['is_male'].value_counts()
```

```python
# Drop columns with more than 30% missing values
threshold = 0.3
cols_to_drop = [col for col in df.columns if df[col].isnull().mean() > threshold]
df.drop(columns=cols_to_drop, inplace=True)
```
